## Run your demo and wait for 3min

```python
from IPython.display import HTML
from base64 import b64encode
command = f"python ./code/demo.py --infer_type 'hubert_audio_only' --stage1_checkpoint_path 'ckpts/stage1.ckpt' --stage2_checkpoint_path 'ckpts/stage2_audio_only_hubert.ckpt' --test_image_path '{image_filename}' --test_audio_path '{audio_filename}' --result_path '{folder_name}'"
os.system(command)
image_name = list(uploaded_image.keys())[0].split('.')[0]
audio_name = list(uploaded_audio.keys())[0].split('.')[0]
video_path = f'/content/AniTalker/{folder_name}/{image_name}-{audio_name}.mp4'
video = open(video_path, "rb").read()
video_encoded = b64encode(video).decode('ascii')
HTML(data=f'''
    <video width="600" controls>
        <source src="data:video/mp4;base64,{video_encoded}" type="video/mp4">
    </video>
''')
```
